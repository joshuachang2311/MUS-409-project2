```python
import numpy as np
import soundfile as sf
import librosa
import librosa.display
import librosa.feature as feature
import matplotlib.pyplot as plt
```

```python
array_1 = np.load("array_1.npy")
array_2 = np.load("array_2.npy")
n_samples = array_1.shape[0]
sampling_rate = 44100
```

```python
n_fft = 2048
hop_length = 512

spec_1 = np.abs(librosa.stft(array_1.T, n_fft=n_fft, hop_length=hop_length))
librosa.display.specshow(librosa.amplitude_to_db(spec_1[0], ref=np.max), sr=sampling_rate, x_axis='time', y_axis='linear')
plt.show()
```

```python
n_fft = 2048
hop_length = 512

spec_2 = np.abs(librosa.stft(array_2.T, n_fft=n_fft, hop_length=hop_length))
librosa.display.specshow(librosa.amplitude_to_db(spec_2[0], ref=np.max), sr=sampling_rate, x_axis='time', y_axis='linear')
plt.show()
```

```python
np.save("spec_1", spec_1)
np.save("spec_2", spec_2)
```

```python
n_fft = 2048
hop_length = 512

mel_spec_1 = feature.melspectrogram(y=array_1.T, sr=sampling_rate, n_fft=n_fft, hop_length=hop_length)
librosa.display.specshow(librosa.power_to_db(mel_spec_1[0], ref=np.max), x_axis='time', y_axis='mel', sr=sampling_rate)
plt.show()
```
